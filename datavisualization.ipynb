```python
#matplotlib
import matplotlib.pyplot as plt

#sample data
years = [2018, 2019, 2020, 2021, 2022, 2023]
sales = [120, 150, 170, 160, 180, 200]

#create the plot
plt.plot(years, sales, marker='o', linestyle='-', color='b', label="Sales")

#add labels and title
plt.xlabel("Year")
plt.ylabel("Sales in (USD)")
plt.title("Yearly Sales Growth")

#show the plot
plt.show()
```

```python
#bar chart

#sample data
categories = ['Electronics', 'Clothing', 'Home Decor', 'Books']
sales = [500, 700, 300, 400]

#create the bar chart
plt.bar(categories, sales, color=['blue', 'green', 'red', 'purple'])

#add labels and titles
plt.xlabel('Product Category')
plt.ylabel('Sales (in USD)')
plt.title('Sales by category')

#show the plot
plt.show()
```

```python
import seaborn as sns 
import pandas as pd

#sample dataset
data = {'category': ['Electronics', 'Clothing', 'Home Decor', 'Books'],'sales': [500, 700, 300, 400]}
df = pd.DataFrame(data)

#create a seaborn plot
sns.barplot(x='category', y='sales', data=df, palette='coolwarm')

#add title
plt.title("sales by category (seaborn)")

#show the plot
plt.show()
```

```python
import numpy as np

#generate random data
data = np.random.randn(1000)

#create a histogram
plt.hist(data, bins=30, color='blue', edgecolor='black')

#add labels and titles
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title('histogram of random data')

plt.show()
```

```python
x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x, y, color="red")

plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Scatter plot of random data')

plt.show()
```
